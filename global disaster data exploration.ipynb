# Global Disaster Data Exploration
In this section, I will explore the dataset of global disasters from 1970 to 2020. The dataset is from Centre for Research on the Epidemiology of Disasters. This data exploration would allow me gain a general idea of how much disasters there were in 50 years by the numbers of columns and rows as well as bar plot. For this training, I used .shape, .info, .head(), .tail(), .plot(), and .value_counts().

## Import geopandas
First of all, I import geopandas and name it as gpd. 

```python
import geopandas as gpd
```

Next, name this file "international" because this is the dataset of international natural disasters from 1970 to 2020.

```python
international = gpd.read_file('disasters_international_50years.csv')
```

## How does the data look like?

### How does the first and last 10 data on the dataset look like?
I use .head() to show the first 10 data on the dataset.

```python
international.head(10)
```

Since these are first 10 data, the year are all 1970. It seems that Hydrological and Meteorological disasters are common in these first 10 data.

Next, I use .tail() to see the last 10 data on the dataset. 

```python
international.tail(10)
```

These last 10 data are data of year of 2020, and most of them are Hydrological disasters.

### Get more information about the dataset

".info()" allows us to see the data types for each column and some detailed information.

```python
international.info()
```

According to this operation, there are 44 columns and 14300 entries in the dataset. 

### Columns and Rows

You can also find how many columns and rows there by using ".shape".

```python
international.shape
```

As it is mentioned, there are 14300 rows and 44 columns.

Now, I use ".columns.to_list()" to see what these 44 columns are.

This dataset has various columns, including country name, disaster subgroup, and total death number.

### Get information about the number of natural disasters in each country
To get the infromation about how many disasters have happened in each country for 50 years, I need to get values for a "country" column.
To do so, I put "country" into "international []" and add ".value_counts()" in order to count how many disasters each country have. 

```python
international['Country'].value_counts()
```

Next, save this code as a variable named "country_count."

```python
country_count = international['Country'].value_counts()
country_count
```

Then, I put this data into a dataframe.

```python
country_count = country_count.reset_index()
country_count
```

You can find what columns there are by using ".columns.to_list()."

```python
country_count.columns.to_list()
```

After creating a new dataframe, I rename column from "index" to "country" and "Country" to "count."

```python
country_count.columns = ['country', 'count']
```

```python
country_count
```

### Visualizing the data
In this section, I use bar graph to visualize the number of natural disaster in each country. The following image is the plot that shows the number of natrual disasters from data id 0 to 226. 

```python
country_count.plot()
```

Now, I create a bar plot and define the name of x and y axis as well as title.

```python
country_count.plot.bar(x = 'country', y = 'count', title = 'Number of natural disaster in each country')
```

Although I was able to create a bar plot of all country, the name of countries are overlapped each other. 
Therefore, I decided to create a bar plot of the top 10 country that has natural diasters by usuing ".head()." 

```python
country_count.head(10)
```

And put this data into a dataframe.

```python
top10_country_count = country_count.head(10)
top10_country_count
```

This is the dataframe of top 10 country with a lot of natural disasters.
Finally, I use ".plot.bar" to recreate a bar plot which shows the number of natural disaster in top 10 country.

```python
top10_country_count.plot.bar(x = 'country', y = 'count', title = 'Number of natural disaster in top 10 country')
```

More than half of these countries are in Asia, and most of them are developing countries.
